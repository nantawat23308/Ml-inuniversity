```python
from sklearn.datasets import make_blobs
X,Y = make_blobs(n_samples =50, centers=2,random_state=0,cluster_std=0.60)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=Y,cmap=plt.cm.Paired, edgecolors= 'black',s=20)
```

```python
from sklearn.linear_model import SGDClassifier
clf=SGDClassifier(loss = 'hinge', alpha=0.001, max_iter = 200)
clf.fit(X,Y)
```

```python
import numpy as np
xx = np.linspace(-1,5,10)
yy = np.linspace(-1,5,10)
X1,X2 = np.meshgrid(xx,yy)
```

```python
Z = np.empty(X1.shape)
```

```python
for (i,j), val in np.ndenumerate(X1):
  x1 = val
  x2 = X2[i,j]
  p = clf.decision_function([[x1,x2]])
  Z[i,j] = p[0]
plt.contour(X1,X2,Z,levels=[0.0],colors = 'b',linstyle='solid')
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,cmap=plt.cm.Paired, edgecolor = 'black',s=20)
plt.contour(X1,X2,Z,levels=[0.0],colour='b',linstyle='soild')
```
