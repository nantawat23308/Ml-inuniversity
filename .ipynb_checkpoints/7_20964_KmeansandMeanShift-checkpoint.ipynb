```python
import pandas as pd
from pandas import read_csv
df = read_csv('data2cluster.csv',delimiter=',',header=None)
```

```python
X =df.values
```

```python
X[0:10,:]
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1])
plt.show
```

```python
from sklearn.cluster import MeanShift
ms = MeanShift()
ms.fit(X)
```

```python
labels = ms.labels_
```

```python
labels[0:10]
```

```python
cluster_centres = ms.cluster_centers_ #return from function ms
```

```python
cluster_centres.shape
```

```python
import numpy as np
colors = np.array(['blue','red'])
plt.scatter(X[:,0],X[:,1],c=colors[list(labels)])
plt.scatter(cluster_centres[:,0],cluster_centres[:,1],c='black') #process build clustering
```

```python
#compare Kmean vs MeanShift
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=2,n_jobs=-1)
km.fit(X)
```

```python
labels2=km.labels_
cluster_centres2 = km.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1],c=colors[list(labels2)])
plt.scatter(cluster_centres2[:,0],cluster_centres2[:,1],c='black')
```
